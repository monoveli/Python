```python
'''
Prova 2 - Estrutura de Dados

Aluna: Mônica Amaral Novelli

nUSP: 11810453

'''
# ------------------------------------------------------------------------------

# O código seguinte foi construído  baseado no Código em Python - Árvore binária
# disponibilizado  no  e-Disciplinas  para a aula 11,  e no material  de  Lowest
# Common Ancestor  do site  towardsdatascience, que pode ser encontrado no link:

# https://towardsdatascience.com/binary-tree-and-lowest-common-ancestor-58eddd433ac

# Além disso, foram utilizados de referências os trabalhos anteriores - em espe-
# cífico, as listas 8 e 9.

# Alguns comentários de códigos anteriores foram reaproveitados.

# ------------------------------------------------------------------------------
# Função de implementação da árvore binária, parcialmente já montado na aula 11.

# Implementa o TAD AB (Arvore Binaria).

# ------------------------------------------------------------------------------
# Classe que implementa uma Arvore Binaria.
# Os atributos desta classe são:
# valor:    valor armazenado na raiz da arvore;
# filhoEsq: filho da esquerda;
# filhoDir: filho da direita.

# ------------------------------------------------------------------------------

class ArvoreBinaria:


# ------------------------------------------------------------------------------
# Construtor. Armazena um valor em "valor", o filho da direita em "filhoDir" e o
# filho da esquerda em "filhoEsq".

    def __init__(self, valor):
        self.valor = valor
        self.filhoEsq = None
        self.filhoDir = None

# ------------------------------------------------------------------------------
# Monta string para impressao da arvore.

    def __str__(self, nivel=0):
        if (self == None):
            return
        else:
            saida = "  "*nivel + str(self.valor) + "\n"
            if (self.filhoEsq != None):
                saida += self.filhoEsq.__str__(nivel+1)
            if (self.filhoDir != None):
                saida += self.filhoDir.__str__(nivel+1)
            return saida

# ------------------------------------------------------------------------------
# Função adicionada para que possamos retirar os nós com '*' adicionados sem al-
# terar a estruturação da árvore. Se retirássemos esses nós na função de cons-
# tutora, então os nós 'vazios' seriam preenchidos por valores da lista, em vez
# de permanecerem vazios.

    def retira_nos(self):
        if self.filhoEsq:
            if self.filhoEsq.valor == '*':
                self.filhoEsq = None
            else:
                self.filhoEsq.retira_nos()

        if self.filhoDir:
            if self.filhoDir.valor == '*':
                self.filhoDir = None
            else:
                self.filhoDir.retira_nos()

# ------------------------------------------------------------------------------
# Cria um no com valor "val_filho" e o insere como filho esquerdo.

    def insereEsq(self,val_filho):
        filho = ArvoreBinaria(val_filho)
        self.filhoEsq = filho


# ------------------------------------------------------------------------------
# Cria um no com valor "val_filho" e o insere como filho direito.

    def insereDir(self,val_filho):
        filho = ArvoreBinaria(val_filho)
        self.filhoDir = filho

# ------------------------------------------------------------------------------
# Usa percurso pre-ordem para imprimir todos os elementos da arvore.

    def preOrdem(self):
        print(self.valor)

        if (self.filhoEsq != None):
            self.filhoEsq.preOrdem()

        if (self.filhoDir != None):
            self.filhoDir.preOrdem()

# ------------------------------------------------------------------------------
# Usa percurso em-ordem para imprimir todos os elementos da arvore.

    def emOrdem(self):
        if (self.filhoEsq != None):
            self.filhoEsq.emOrdem()

        print(self.valor)

        if (self.filhoDir != None):
            self.filhoDir.emOrdem()


# ------------------------------------------------------------------------------
# Usa percurso pos-ordem para imprimir todos os elementos da arvore.

    def posOrdem(self):
        if (self.filhoEsq != None):
            self.filhoEsq.posOrdem()

        if (self.filhoDir != None):
            self.filhoDir.posOrdem()

        print(self.valor)


# ------------------------------------------------------------------------------
# Busca "valor" na arvore, usando percurso pre-ordem.

    def busca(self,valor):
        if (self.valor == valor):
            return True

        if (self.filhoEsq != None and self.filhoEsq.busca(valor) == True):
            return True

        if (self.filhoDir != None and self.filhoDir.busca(valor) == True):
            return True

        return False

# ------------------------------------------------------------------------------

# ------------------------------------------------------------------------------

# Funções para a construção da árvore fora da classe.

# ------------------------------------------------------------------------------
# Função auxiliar para que possamos inserir elementos em uma árvore.

def insere_na_arvore(raiz, valor):
   lista = []
   lista.append(raiz)
   # Loop para rodar todos os valores da lista
   while (len(lista)):
      raiz = lista[0]
      lista.pop(0)
      if (not raiz.filhoEsq):
        # Inicialmente, a ideia era alterar os nós vazios no próprio construtor,
        # entretanto, como explicado anteriormente,  mudou a estruturação da ár-
        # vore errôneamente. Entretanto, foram mantidas as funções para referen-
        # ciação apenas adicionando '*' na árvore, mantendo seu valor.
         if valor is not '*':
            raiz.filhoEsq = ArvoreBinaria(valor)
         else:
            raiz.filhoEsq = ArvoreBinaria('*')
         break
      else:
         lista.append(raiz.filhoEsq)
         if (not raiz.filhoDir):
            if valor is not '*':
               raiz.filhoDir = ArvoreBinaria(valor)
            else:
               raiz.filhoDir = ArvoreBinaria('*')
            break
         else:
            lista.append(raiz.filhoDir)

# ------------------------------------------------------------------------------
# Função principal para que possamos inserir elementos em uma árvore, utilizando
# de uma lista para isso.

def construtor(valores):
   arvore = ArvoreBinaria(valores[0]) # primeiro elemento da lista é a raiz
   for item in valores[1:]:
      insere_na_arvore(arvore, item) # insere itens utilizando da auxiliar
   return arvore

# ------------------------------------------------------------------------------
# Função para verificar se existe um caminho para 'encontrar' um nó.
# Retorna um booleano,  que utilizaremos  nas próximas funções para verificar se
# é possível encontrar esse caminho ou não.

def encontra_caminho(raiz, caminho, n):
    if raiz is None:
      return False

    caminho.append(raiz.valor)

    if raiz.valor == n :
        return True

    if ((raiz.filhoEsq != None and encontra_caminho(raiz.filhoEsq, caminho, n)) or (raiz.filhoDir!= None and encontra_caminho(raiz.filhoDir, caminho, n))):
        return True

    caminho.pop()
    return False


# ------------------------------------------------------------------------------
# Essa função encontra o nó que faz a conexão principal entre os setores a serem
# comunicados.  Ou seja,  encontra o menor ancestral em comum entre eles,  o que
# seria a 'raiz' da comunicação.

def conexao(raiz, n1, n2):


    caminho1 = []
    caminho2 = []

    # utilizando o booleano para ver se existe um caminho possível para esses
    # nós ou não
    if (not encontra_caminho(raiz, caminho1, n1) or not
        encontra_caminho(raiz, caminho2, n2)):
        return -1

    # encontrando a posição do menor ancestral comum por meio de um contador
    i = 0
    while (i < len(caminho1) and i < len(caminho2)):
        if caminho1[i] != caminho2[i]:
            break
        i += 1
    return caminho1[i-1] # busca a posição do menor ancestral comum e o retorna

# ------------------------------------------------------------------------------
# Encontra o caminho entre os dois nós a serem comunicados,  e retorna uma lista
# com esse caminho.

def caminho(raiz, n1, n2):

    caminho1 = []
    caminho2 = []

    # utilizando o booleano para ver se existe um caminho possível para esses
    # nós ou não
    if (not encontra_caminho(raiz, caminho1, n1) or not
        encontra_caminho(raiz, caminho2, n2)):
        return -1

    l = []

    # verifica intersecção entre os caminhos
    for item in caminho1:
        if (item in caminho2) and ((item != n1) and (item != n2)):
            l.append(item)

    # retira os caminhos individuais da nossa lista
    for item in l:
        caminho1.remove(item)
        caminho2.remove(item)

    # monta nossa lista do caminho final e o retorna
    else:
        caminho = []
        caminho += caminho1[::-1]
        caminho.append(l[-1])
        caminho += caminho2
        return caminho

# ------------------------------------------------------------------------------
# Função para printar o caminho individual da raiz até um nó

def printcaminho(raiz, x):

    l = []

    if (encontra_caminho(raiz, l, x)):
        for i in range(len(l) - 1):
            print(l[i], end = ' -> ')
        print(l[len(l) - 1])

    else:
        print("n existe caminho")

# ------------------------------------------------------------------------------
# Função para printar uma lista como um caminho

def printlista(lista):

    for i in range(len(lista) - 1):
        print(lista[i],end = ' -> ')
    print(lista[len(lista) - 1])

# ------------------------------------------------------------------------------

# Programa principal

# ------------------------------------------------------------------------------

lista = []

print('Digite a hierarquia da Empresa:\n')

# Recebendo a entrada:
while True:
    entrada = input()
    if entrada == 'X':
        break
    lista += entrada.split(' : ')

# Montando e imprimindo a árvore:
arvore = construtor(lista)

arvore.retira_nos()
print()
print(arvore)

caminho0 = []

# Recebendo os setores que se comunicarão:
print('''\nDigite os setores que se comunicarão, separados por uma vírgula (' , '):\n
''')

n1, n2 = input().split(' , ')

print('\nEsse é o setor que fará a conexão entre principal entra as duas áreas a serem comunicadas:\n')

canal = conexao(arvore, n1, n2)

# Imprime o menor ancestral comum entre eles:
print(canal)

# Saída Opcional:

# Imprime os caminhos de cada setor, partindo da raiz da árvore.

#print('Esses são os caminhos completos de cada setor:')

#printcaminho(arvore,n1)
#printcaminho(arvore,n2)

print('\nEsse é o caminho que ocorrerá a comunicação:\n')
# Imprime o caminho entre eles:
lista = caminho(arvore, n1, n2)
printlista(lista)
```
